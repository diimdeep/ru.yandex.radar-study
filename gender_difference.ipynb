```python
import pandas as pd
```

```python
df_f = pd.read_csv('18-34_MO_rich_female.csv')
df_m = pd.read_csv('18-34_MO_rich_male.csv')
```

```python
cols = ['name', 'url', 'Тематики ресурса',
       'Тип ресурса', 'Данные Метрики', 'users',
       'Посетители (браузер)', 'time',
       'Доля пользователей приложения', 'users_']

df_m.columns = cols
df_f.columns = cols
```

```python
df_mm = df_m[['name', 'url', 'time', 'users']]
df_ff = df_f[['name', 'url', 'time', 'users']]
df_mm = df_mm.set_index('url')
df_ff = df_ff.set_index('url')
```

```python
df = df_mm.join(df_ff, how='outer', lsuffix='_m', rsuffix='_f')#[['Название ресурса', 'URL-адрес ресурса', 'Дневная аудитория', 'URL-адрес ресурса']]
df
```

```python
# df = df.sort_values('users_f', ascending=False)
# df = df.iloc[10:]
```

```python
df['diff'] = df.apply(lambda x: x.users_f - x.users_m, axis=1)
#df['diff'] = df.apply(lambda x: x['diff'] / df['diff'].mean(), axis=1)
```

```python
df = df.sort_values('diff', ascending=False)
df_female = df.iloc[:60][['users_m', 'users_f', 'diff']]
df_female.columns = ["Муж аудитория", "Жен аудитория", "Разница"]
df_female.style.bar(subset=["Муж аудитория", "Жен аудитория", "Разница"])
```

```python
df = df.sort_values('diff', ascending=True)
df_male = df.iloc[:60][['users_m', 'users_f', 'diff']]
df_male['diff'] = df_male.apply(lambda x: abs(x['diff']), axis=1)
df_male.columns = ["Муж аудитория", "Жен аудитория", "Разница"]
df_male.style.bar(subset=["Муж аудитория", "Жен аудитория", "Разница"])
```

```python
df['diff_abs'] = df.apply(lambda x: abs(x.users_f - x.users_m), axis=1)
#df['diff'] = df.apply(lambda x: x['diff'] / df['diff'].mean(), axis=1)
df = df.sort_values('diff_abs', ascending=True)
df.iloc[:60][['users_m', 'users_f', 'diff_abs']].style.bar(subset=['users_f', 'users_m', 'diff_abs'])
```
